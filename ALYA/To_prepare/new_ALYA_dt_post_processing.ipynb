Objectives are defined as follows:
1. Plot $dt$ as a function of $t$ to establish a fixed timestep for subsequent analysis.
2. Visualize the distribution of $dt$.


```python
import matplotlib.pyplot as plt
```

```python
def extract_data(file_path):
    times = []
    time_steps = []
    with open(file_path, 'r') as file:
        for line in file:
            if "Current time step dt" in line:
                dt_value = float(line.split('=')[1].strip())
            elif "Current time t" in line:
                t_value = float(line.split('=')[1].strip())
                # Append both dt and t only after both have been found
                times.append(t_value)
                time_steps.append(dt_value)
    return time_steps, times
```

```python
# Example usage:
file_path = 'longer_run_vtk/channel.log'
time_steps, times = extract_data(file_path)
```

```python
min(time_steps)
```

```python
def plot_dt_vs_t(time_steps, times):
    plt.figure(figsize=(10, 5))
    plt.plot(times, time_steps,  linestyle='-')
    plt.title('Current time step dt vs Current time t')
    plt.xlabel('Current time t')
    plt.ylabel('Current time step dt')
    plt.yscale('log')
    plt.grid(True)
    plt.show()

plot_dt_vs_t(time_steps, times)
```

```python
def plot_dt_distribution(time_steps):
    plt.figure(figsize=(10, 5))
    plt.hist(time_steps, bins=20, alpha=0.75)
    plt.title('Distribution of Current time step dt')
    plt.xlabel('Current time step dt')
    plt.ylabel('Frequency')
    plt.xscale('log')
    plt.grid(True)
    plt.show()
    
plot_dt_distribution(time_steps)
```
